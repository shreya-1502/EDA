```python
mean1 = df['Price'].mean()  #displays mean of the column 
print ('Mean :' + str(mean1))

sum1 = df['Price'].sum()  #displays sum of the column 
print ('Sum :' + str(sum1))

max1 = df['Price'].max()  #displays maximum number of the column 
print ('Maximum :' + str(max1))

min1 = df['Price'].min()  #displays minimum number of the column 
print ('Minimum :' + str(min1))

count1 = df['Price'].count()  #displays count of the column 
print ('Count :' + str(count1))

median1 = df['Price'].median() #displays median of the column 
print ('Median :' + str(median1))

std1 = df['Price'].std() #displays standard deviation of the column 
print ('Standard Deviation :' + str(std1))

var1 = df['Price'].var()  # displays variance of the column 
print ('Variance :' + str(var1))
```

```python
count1 = df['Price'].count()  #displays count of the column 
print ('Count :' + str(count1))
sum1 = df['Price'].sum()  #displays sum of the column 
print ('Sum :' + str(sum1))
print('-----------')
average = sum1/count1  # average of the column 
print ('Average :' + str(average))
```
